```python
from clustering import ClustarGenerator
from vectrizer import ContentVectorizer

import numpy as np

cv = ContentVectorizer()
cg = ClustarGenerator()
```

```python
data_path = "data/sample_doc.txt"
data_list = []
with open(data_path, "r") as f:
    data_list = [line.replace("\n", "") for line in f.readlines()]
len(data_list)
```

```python
vecotor_list = np.array([cv.vectorize(data)[1] for data in data_list])
len(vecotor_list)
```

```python
cluster_result = cg.generate(title_list=data_list, 
            vector_list=vecotor_list, 
            vector_size=100, 
            n_cluster=10)
```
