# ベクトル化を行うモジュール

+ ginzaを用いて簡単なベクトル化を行う
+ 機能としては以下の通り
  + 粒度の違うベクトル化
    + ドキュメント単位
    + 文章単位
    + 単語単位
  + ファイル出力用のjsonを出力
  + 学習済みモデルに差し替え

```python
from vectrizer import ContentVectorizer
cv = ContentVectorizer()
```

```python
text = """
【Material】肌離れがよく、上品でツヤのあるきれい目な素材。通勤にもおすすめな素材感です。
【Design】表にはシアーなシフォン素材、裏には透けないカットソー素材を使用しているので、インナーを気にせず1枚でシアーなスタイルが完成。
後ろのスリットデザインは動きが出てこなれ感がでます。【Styling  point】シアー感があり、着るだけでトレンドのスタイリングになるトップス。
今の気分は、トレンドのミントやパープル、夏らしい柄のボトムはもちろん、シンプルなボトムに合わせたクリーンな雰囲気にしても○。
透け:本体なし、袖部分あり裏地:なし伸縮性:あり光沢感:なし生地の厚さ:薄手ポケット:なし洗濯方法:手洗い可※アウトレット商品のため、タグにサービスプライス商品の記載がございます。【BRAND】AER。
"""
# ドキュメント単位
# 返り値は(doc class, vector)のタプル
cv.vectorize(text)
```

```python
# 文章単位
# 返り値は[(sent class, vector)]のタプルのリスト
cv.sent_vectorize(text)[0]
```

```python
# 単語単位
# 返り値は[(token class, vector)]のタプルのリスト
cv.word_vectorize(text)[0]
```

```python
cv.ginza.vocab.vectors.shape
```

## 学習済みベクトルへの差し替え

+ word2vecでsaveした形式であればベクトルの差し替えが可能
    + 差し替え前: 300次元
    + 差し替え後: 100次元

```python
#cv = ContentVectorizer() # 初期化時でも可能
model_path = "model/awg_word2vec.model"
cv.load_model(model_path)
cv.ginza.vocab.vectors.shape
```

## ファイルへ出力

+ jsonl形式への変換を行う
    + 返り値は [json] でjsonのリスト
+ 保存自体は呼び出し側で行う

```python
# 基本
cv.dump_vector(text,title="サンプルベクトル", vector_type="doc")
```

```python
# 保存例
save_path = "result/sample.jsonl"
with open(save_path, "w") as f:
    dump_json = cv.dump_vector(text,title="サンプルベクトル", vector_type="doc")
    f.write("\n".join(dump_json))
```

```python
# 文章単位
cv.dump_vector(text, vector_type="sent")[:2]
```

```python
# 単語単位
cv.dump_vector(text, vector_type="word")[:2]
```
